```python
rdf = Exp.export_results_df()
```

```python
sids = rdf["Simulation Id"].unique()
print(sids)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, a = plt.subplots(2,10,figsize=(20,4))

for sid in sids:
    sdf = rdf[rdf["Simulation Id"]==sid].copy()

    sdf['Ax'] = sdf.Alice.apply(lambda p: p.x)
    sdf['Ay'] = sdf.Alice.apply(lambda p: p.y)
    sdf['Bx'] = sdf.Bob.apply(lambda p: p.x)
    sdf['By'] = sdf.Bob.apply(lambda p: p.y)

    i1 = sid%2
    i2 = int(sid/2)
    a[i1][i2].plot(sdf.timestep,sdf.Ax,
    sdf.timestep,sdf.Ay,
    sdf.timestep,sdf.Bx,
    sdf.timestep,sdf.By,'.')
```
