```python
df = Sim.results2df(dropIntermediates=True)
```

```python
df.head(20)
```

```python
df["Xa"] = df.Alice.apply(lambda p: p.x)
df["Ya"] = df.Alice.apply(lambda p: p.y)
df["Xb"] = df.Bob.apply(lambda p: p.x)
df["Yb"] = df.Bob.apply(lambda p: p.y)
```

```python
import matplotlib.pyplot as plt
plt.plot(df.Xa.values, df.Ya.values, '+')
plt.plot(df.Xb.values, df.Yb.values,'x')
plt.legend(['Alice', 'Bob'])
plt.xlabel("x position")
plt.ylabel("y position")
```

```python
fig,a =  plt.subplots(2,1)
a[0].plot(df.timestep.values, df.Xa.values, df.timestep, df.Xb.values)
a[1].plot(df.timestep.values,df.Ya.values, df.timestep.values, df.Yb.values)
plt.xlabel("time")
plt.legend(['Alice', 'Bob'])
```

```python
import numpy as np
df["distance"] = np.sqrt((df.Xa-df.Xb)**2 + (df.Ya-df.Yb)**2)
```

```python
plt.plot(df[df['It']=="Alice"].timestep.values, df[df['It']=="Alice"].distance, '.')
plt.plot(df[df['It']=="Bob"].timestep.values, df[df['It']=="Bob"].distance, '.')
plt.legend(['Alice It', 'Bob It'])
```
